```python
# Game Coding Start Here
# Gobal Variables
import random
suits = ("Hearts", "Diamonds", "Spades", "Clubs")
ranks = ("Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace")
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10,
          'King':10, 'Ace':11}

# Declring game is on
playing = True

# Objects for our games starts here

# Card class start here
class Cards():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + " of " + self.suit
# Card class end here

# Deck class start here
class Deck():
    
    def __init__(self):
        # start with an empty list
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Cards(suit,rank))
    
    # String Representation of Deck
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n' + card.__str__()
        return "The deck has: "+deck_comp
    
    # Shuffle the deck cards
    def shuffle(self):
        random.shuffle(self.deck)
    
    # Garb a deck card
    def deal(self):
        single_card = self.deck.pop()
        return single_card
# Deck class end here
    
# Hand class start here
class Hand():
    
    def __init__(self):
        # start with an empty list 
        self.cards = [] 
        # value attribute to sum the cards value start with zero value
        self. value = 0
        # attribute for track of aces as this version of game has some especially rules
        self.aces = 0
    
    def add_card(self,card):
        # card passed in is from Deck.deal() basically the single card(suit,rank)
        self.cards.append(card)
        self.value += values[card.rank]
        
        # track aces
        if card.rank == 'Ace':
            self.aces += 1
    
    def adjust_for_ace(self):
        
        # when value > 21 and we have an ace then change the ace value to 1 instead of 11
        while self.value > 21 and self.aces > 0:
            self.value -= 10
            self.aces -= 1        
# Hand class end here

# Chips class start here
# this object track players starting chips, bets and ongoing wins
class Chips():
    
    def __init__(self):
        # start valus of chips and bets
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
# Chips class end here

# Objects for our games end here 

# Function for our games starts here

# Function for game bet start here
def take_bet(chips):
    
    while True:
        
        try: 
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print('Sorry please provide an integer')
        else:
            if chips.bet > chips.total:
                print('Sorry, you do not have enough chips! You have: {}'.format(chips.total))
            else:
                break
# Function for game bet end here

# Function to hit a card start here
def hit(deck,hand):
    
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
# Function to hit a card end here

# Function to prompting hit or stay start here
def hit_or_stay(deck,hand):
    
    # to control upcoming while loop
    global playing
    
    while True:
        x = input("Would you like to Hit or Stay? Enter 'h' for Hit and 's' for Stay")
        
        if x[0].lower() == 'h':
            hit(deck,hand)
        
        elif x[0].lower() == 's':
            print('Player Stay. Dealer is playing.')
            playing = False
        
        else:
            print('Sorry I did not understand that, please enter h or s only!')
            continue
        
        break
# Function to prompting hit or stay end here

# Function to Display cards start here
# 1st function for displaying cards at starting
def show_some(player,dealer):
    
    # show all dealer cards
    # this version game has one card face down and one card face up at starting
    print("\n Dealer's Card:")
    print("First card is hidden!")
    print(dealer.cards[1])
    
    # show all player cards at starting only 2 cards
    print("\n Player's Card:")
    for card in player.cards:
        print(card)

# this function for displaying cards after 1st round
def show_all(player,dealer):    
    # show all dealer cards
    print("\n Dealer's Card:")
    for card in dealer.cards:
        print(card)
        
    # calculate and display value (Q+K==20) for dealer
    print(f"Value of Dealer's card is : {dealer.value}")
    
    # show all player cards
    print("\n Player's Card:")
    for card in player.cards:
        print(card)
    
    # calculate and display value (Q+K==20) for player
    print(f"Value of Player's card is : {player.value}")
# Function to Display cards end here

# Function to handle game scenarios start here
def player_busts(player,dealer,chips):
    print("PLAYER BUST!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("PLAYER WINS!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("DEALER BUST! PLAYER WINS")
    chips.lose_bet()

def dealer_wins(player,dealer,chips):
    print("DEALER WINS!")
    chips.win_bet()
    
def push(player,dealer):
    print("Dealer and Player Tie! PUSH")
# Function to handle game scenarios end here

# Function for our games end here

# Logic for the game start here
while True:
    
    # Starting massage
    
    print('WELCOME TO BLACKJACK GAME')
    # Create & shuffle the deck, deal two cards to each player and dealer
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    # Set up the player's chips
    player_chips = Chips()
    
    # Prompt the player for their bet
    take_bet(player_chips)
    
    # show cards (remember one card of dealer is hidden)
    show_some(player_hand,dealer_hand)
    
    # call variable from hit_or_stay function
    while playing:
        
        # Prompt for player to Hit or Stay
        hit_or_stay(deck,player_hand)
        
        # show cards (remember one card of dealer is hidden)
        show_some(player_hand,dealer_hand)
        
        # check if player exceeds 21 run player busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            
            break
        
        # if the player has not busted, play dealer hand until dealer value < player value
        if player_hand.value <= 21:
            
            while dealer_hand.value < player_hand.value:
                hit(deck,dealer_hand)
        
        # show all cards 
        show_all (player_hand,dealer_hand)
        
        # different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
        else:
            push(player_hand,dealer_hand)
        
    # inform player about the  total chips
    print('\n Player total chips are at {}'.format(player_chips.total))

    # Ask to play again
    new_game = input('Would you like to play another hand? y/n')

    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print('Thank you for playing!')

        break    
# Logic for the game end here

# Game codes ends here
```
